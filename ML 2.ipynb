# Assignment no 2

Classify the email using the binary classification method.
EmailSpam detection has two states:
1. Normal State – Not Spam
2. Abnormal State – Spam     

Use K-Nearest Neighbors and Support Vector Machine for classification.
Analyze their performance.


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier  
from sklearn import metrics
```

```python
df = pd.read_csv('emails.csv')
df
```

```python
df.shape
```

```python
# Check if any column in the DataFrame has missing values (True = missing values exist)
df.isnull().any()
```

```python
df.drop(columns='Email No.', inplace=True)
df
```

```python
df.columns
```

```python
# Retrieve the unique values from the 'Prediction' column in the DataFrame
df.Prediction.unique() 
```

```python
df['Prediction'] = df['Prediction'].replace({0:'Not spam', 1:'Spam'})
```

```python
df
```

## KNN

```python
# Separate features (X) by dropping 'Prediction' column and target (Y) by selecting 'Prediction' column

X = df.drop(columns='Prediction',axis = 1)
Y = df['Prediction']
```

```python
X.columns
```

```python
Y.head()
```

```python
# Split data into training (80%) and testing (20%) sets with a fixed random seed for reproducibility

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
# Initialize and train a KNN classifier with 7 neighbors, then make predictions on the test set

KN = KNeighborsClassifier
knn = KN(n_neighbors=7)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
print("Prediction: \n")
print(y_pred)
```

```python
# Accuracy

M = metrics.accuracy_score(y_test,y_pred)
print("KNN accuracy: ", M)
```

## SVM Classifier

```python
# Train an SVC model with regularization parameter C=1 and make predictions on the test set

model = SVC(C = 1) 
model.fit(x_train, y_train)
y_pred = model.predict(x_test)      
```

```python
# [[TN, FP],
#  [FN, TP]]

kc = metrics.confusion_matrix(y_test, y_pred)
print("SVM accuracy: ", kc)
```
